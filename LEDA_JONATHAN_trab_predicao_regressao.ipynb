Ao analisar a distribuição de freqüências de uma variável quantitativa devemos verificar basicamente três características:
    
> 1ª característica: Tendência Central
    
É a classe de maior freqüência, chamada de classe modal. No histograma, esta classe corresponde àquela com barra mais alta ("pico"). No caso abaixo, temos um histograma que pode ser considerado bimodal pois indica a existência de dois grupos, com valores centrados em dois pontos diferentes do eixo de valores.
    
OBS: No primeiro gráfico de histograma, devido ao baixo número de bins (barras), os dois pontos centrais não estão tão claros, porém ao aumentar o número de bins, como no segundo gráfico, podemos ver com mais clareza a tendência bimodal. 

```python
# Gráfico de distribuição de frequência da variável ValuePerSqFt
histogramaPLT(housing['ValuePerSqFt'], 50)
```

```python
# Gráfico de distribuição de frequência da variável ValuePerSqFt
histogramaSNS(housing['ValuePerSqFt'],500)
```

> 2º característica: Variabilidade

Indica a variação, ou seja, a amplitule dos valores de cada ocorrência. No caso abaixo podemos notar que apesar dos valores oscilarem de, aproximadamente 10 a 399 dólares por pés quadrado, a concentração de variação está entre 25 e 275. 
 
> 3º característica: Forma

A terceira característica corresponde à observação de simetria ou assimetria do gráfico. No caso acima, temos uma tendência de assimetria, pois existe uma maior concentração à esquerda e a formação de uma cauda mais longa à direita (histograma assimétrico à direita).

### 8. **(0,5 ponto) Construa os modelos de regressão:**

    - house1:ValuePerSqFt~Units+SqFt+YearBuilt+Income+Boro
    - house2:ValuePerSqFt~Units*SqFt+YearBuilt+IncomePerSqFt+Income+Boro
    - house_log1:ValuePerSqFt~log⁡(Units)+log⁡(SqFt)+Boro
    - house_log2:log⁡(ValuePerSqFt)~log⁡(Units)+log⁡(SqFt)+Boro

```python
# importando o pacote utilizado para a modelagem
import statsmodels.formula.api as smf
import numpy as np
```

```python
# criando as variáveis aplicando log
housing['Units_log'] = np.log(housing['Units'])
housing['SqFt_log'] = np.log(housing['SqFt'])
housing['ValuePerSqFt_log'] = np.log(housing['ValuePerSqFt'])
```

```python
# fitando os quatro modelos sugeridos
house1 = smf.ols('ValuePerSqFt ~ Units + SqFt + YearBuilt + Income + C(Boro)', data=housing).fit()
house2 = smf.ols('ValuePerSqFt ~ Units*SqFt + YearBuilt + IncomePerSqFt + Income + C(Boro)', data=housing).fit()
house_log1 = smf.ols('ValuePerSqFt ~ Units_log + SqFt_log + C(Boro)', data=housing).fit()
house_log2 = smf.ols('ValuePerSqFt_log ~ Units_log + SqFt_log + C(Boro)', data=housing).fit()
```

Não foi feita nenhuma modificação para lidar com a variável dummy Boro porque a biblioteca Statsmodels já lida muito bem com isso. Basta indicar com o C que a variável é categórica que a biblioteca já entende e faz o que é necessário para modelar. O comando modelo.summary() mostra as principais estatísticas do modelo.

```python
house1.summary()
```

### 9. **Compare os modelos usando AIC(modelo1, modelo2, modelo3,modelo4)**

Obs.: Um dos critérios para se comparar modelos de regressão é o Akaike Information Criteria AIC. 

```python
# criando um dataframe com os valores do AIC de cada modelo
models_aic = pd.DataFrame()

models_aic['house1'] = [house1.aic]
models_aic['house2'] = [house2.aic]
models_aic['house_log1'] = [house_log1.aic]
models_aic['house_log2'] = [house_log2.aic]


models_aic = models_aic.round(2)
models_aic = models_aic.rename({0:'AIC'})
models_aic
```

Olhando para o AIC, diríamos que o melhor modelo é o house_log2. O AIC é uma medida de qualidade do modelo e quanto menor o seu valor, (em comparação com modelos similares) mais adequado o modelo está aos dados.

Chamamos de Análise dos Resíduos um conjunto de técnicas utilizadas para investigar a adequabilidade de

O resíduo é dado pela diferença entre a variável resposta observada e a variável resposta estimada.
A ideia básica da análise dos resíduos é que, se o modelo for apropriado, os resíduos devem refletir as
propriedades impostas pelo termo de erro do modelo.

O gráfico dos resíduos versus valores ajustados: verifica a homoscedasticidade do modelo, isto é, σ2 constante.

O gráfico dos resíduos versus valores ajustados (valores preditos) é uma das principais técnicas utilizadas para verificar as suposições dos resíduos. 

Para o diagnóstico de heteroscedasticidade, tentamos encontrar alguma tendência no gráfico. Por isso, se os pontos estão aleatoriamente distribuídos em torno do 0, sem nenhum comportamento ou tendência, temos indícios de que a variância dos resíduos é homoscedástica. Já a presença de "funil", por exemplo, é um indicativo da presença de heteroscedasticidade.

**No caso dos gráficos acima, parece que os resíduos de nenhum dos modelos são homocedásticos, pois o valores não ficam aleatoriamente em torno de zero.**

### 11. **(0,5 ponto) Faça o correspondente teste de Hipótese. Qual sua conclusão para cada modelo?**

```python
# teste te Breusch-Pagan para os 4 modelos (teste de homocedasticidade)
from statsmodels.stats.diagnostic import het_breuschpagan

def teste_heterocedasticidade(modelo):
    print('Estatística LM:   ',    het_breuschpagan(modelo.resid, modelo.model.exog)[0])
    print('p-valor teste LM: ',    het_breuschpagan(modelo.resid, modelo.model.exog)[1])
    print('Estatística F:    ',    het_breuschpagan(modelo.resid, modelo.model.exog)[2])
    print('p-valor teste F:  ',    het_breuschpagan(modelo.resid, modelo.model.exog)[3])
    print('')
    
teste_heterocedasticidade(house1)
teste_heterocedasticidade(house2)
teste_heterocedasticidade(house2)
teste_heterocedasticidade(house_log1)
```

Teste de homocedasticidade:
Para o diagnóstico de heteroscedasticidade, tentamos encontrar alguma tendência no gráfico.
Por isso, se os pontos estão aleatoriamente distribuídos em torno do 0, sem nenhum comportamento ou tendência, temos indícios de que a variância dos resíduos é homoscedástica.

 - H0: hipótese de homocedasticidade (variância constante)
 - H1: hipótese de heterocedasticidade (variância não é constante)
 
Considerando alpha = 0.05, observamos que o primeiro e o último modelo são significativos. Sendo assim, podemos rejeitar a hipótese nula (hipótese de homocedasticidade). Nos outros dois modelos, o p-valor foi muito alto e, portanto, não temos evidências para rejeitarmos a hipótese nula.

### 13. **(0,5 ponto) Faça o correspondente teste de hipótese. Qual sua conclusão?**

```python
# teste de Shapiro-Wilk
from scipy.stats import shapiro

def teste_shapiro_wilk(variavel):
    
    stat, p = shapiro(variavel)

    print('Statistics: %.3f' % (stat))
    print('valor-p: %.3f' % (p))

    # interpretação
    alpha = 0.05

    if p > alpha:
        print('A distribuição é normal (não rejeitamos H0).')
    else:
        print('A distribuição não é normal (rejeitamos H0).')
    print('')

# testando a normalidade os resíduos do modelo
teste_shapiro_wilk(house1.resid_pearson)
teste_shapiro_wilk(house2.resid_pearson)
teste_shapiro_wilk(house_log1.resid_pearson)
teste_shapiro_wilk(house_log2.resid_pearson)

# testando as variáveis resposta
teste_shapiro_wilk(housing.ValuePerSqFt)
teste_shapiro_wilk(housing.ValuePerSqFt_log)
```

Em todos os casos, o teste de hipóteses mostra que a distribuição não é normal (o que já era esperado devido à bimodalidade encontrada nos histogramas).

### 14. **(0,5 ponto) Algum dos modelos apresenta problemas de multicolinearidade?**

```python
corr1 = housing[['ValuePerSqFt', 'Units', 'SqFt', 'YearBuilt', 'Income']].corr()
corr1
```

```python
def plot_corr(corr):

    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0,
                square=True, linewidths=.5, cbar_kws={'shrink': .5});
    
plot_corr(corr1)
```

A multicolinearidade é um problema comum em regressões, no qual as variáveis independentes possuem fortes relações lineares.
